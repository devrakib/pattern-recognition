```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```
from random import random
# datasets
x = []
y = []

### First coordinate value
for i in range(25):
  x.append(random())
  y.append(random())

### Third coordinate value
for i in range(25):
  x.append(-random())
  y.append(-random())

### Second coordinate value
for i in range(10):
  x.append(random())
  y.append(-random())

### Fourth coordinate value
for i in range(10):
  x.append(-random())
  y.append(random())

plt.scatter(x, y, color="green")
plt.show()
```

```

c1x = random()
c1y = random()

c2x = -random()
c2y = -random()

plt.scatter(x, y, color="green")
plt.scatter(c1x, c1y, color = "red")
plt.scatter(c2x, c2y, color = "blue")
plt.show()
```

```
# d = sqrt (( c1x - x[i]) ** 2 + ( c1y - y[i]) ** 2 )
d1 = []
d2 = []
for i in range(len(x)):
  dred= ( (( c1x - x[i]) ** 2) + (( c1y - y[i]) ** 2) )
  d1.append(dred)

for i in range(len(x)):
  dred= ( (( c2x - x[i]) ** 2) + (( c2y - y[i]) ** 2) )
  d2.append(dred)

print(d1)
print(d2)
```

```
for i in range(len(x)):
  if d1[i] < d2[i]:
    plt.scatter(x[i], y[i], color = "red")
  else:
    plt.scatter(x[i], y[i], color = "blue")
    
plt.scatter(c1x, c1y, color = "red", marker="X")
plt.scatter(c2x, c2y, color = "blue", marker="X")
plt.show()
```

```
count_red = 0
count_blue = 0
sum_red_x = 0
sum_red_y = 0
sum_blue_x = 0
sum_blue_y = 0
for i in range(len(x)):
  if d1[i] < d2[i]:
    plt.scatter(x[i], y[i], color = "red")
    sum_red_x = sum_red_x + x[i]
    sum_red_y = sum_red_y + y[i]
    count_red = count_red +1
  else:
    plt.scatter(x[i], y[i], color = "blue")
    sum_blue_x = sum_blue_x + x[i]
    sum_blue_y = sum_blue_y + y[i]
    count_blue = count_blue +1

avg_red_x = sum_red_x/count_red
avg_red_y = sum_red_y/count_red

avg_blue_x = sum_blue_x/count_blue
avg_blue_y = sum_blue_y/count_blue

print(avg_red_x) 
print(avg_red_y) 
print(avg_blue_x) 
print(avg_blue_y) 

c1x = avg_red_x
c1y = avg_red_y
c2x = avg_blue_x
c2y = avg_blue_y

plt.scatter(c1x, c1y, color = "red", marker="X")
plt.scatter(c2x, c2y, color = "blue", marker="X")
plt.show()
```
