Precision = TP / TP + FP

Recall = TP / TP + FN

F1 = (2 * Precision * Recall) / (Precision + Recall)

```
#dataset
from random import randint

prediction = []
result = []

for i in range(1000):
  prediction.append(randint(0,1))

for i in range(1000):
  result.append(randint(0,1))

print(prediction)
print(result)
```

```
TN = 0
TP = 0
FP = 0
FN = 0
for i in range(1000):
  if prediction[i] == 0 and result[i] == 0:
    TN += 1
  if prediction[i] == 1 and result[i] == 1:
    TP += 1
  if prediction[i] == 1 and result[i] == 0:
    FP += 1
  if prediction[i] == 0 and result[i] == 1:
    FN += 1
  
precision = TP/(TP + FP)
recall = TP/(TP+ FN)
f1 = (2*precision*recall)/(precision+recall)

print("True Positive = ", TP)
print("True Negative = ", TN)
print("False Positive = ", FP)
print("False Negative = ", FN)
print("")
print("Precision = ", precision)
print("Recall = ", recall) 
print("")
print("F1 score = ", f1)
```
