```
import matplotlib.pyplot as plt

from sklearn import datasets

from sklearn import svm

digits = datasets.load_digits()
print(digits.data)

print(digits.target)
print(digits.images[0])
print(digits.images[1])

clf = svm.SVC(gamma = 0.0001, C = 100)
x = digits.data[10:-1]
y = digits.target[10:-1]

clf.fit(x,y)
print("prediction: ", clf.predict(digits.data[[8]]))
print(len(digits.target))

plt.imshow(digits.images[8], cmap = plt.cm.gray_r )
plt.show()
```

```
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import svm
from random import randint
# datasets
x = []
y = []
coordinate = []
for i in range(-1000, 1000):
  x.append(randint(-1000, 1000))
  y.append(randint(-1000, 1000))

# label the coordinate
#  1st coordinate = 1
#  2nd coordinate = 2
#  3rd coordinate = 3
#  4th coordinate = 4
for i in range(len(x)):
  if x[i] >= 0 and y[i] >= 0:
    coordinate.append(1)
  if x[i] < 0 and y[i] > 0:
    coordinate.append(2)
  if x[i] < 0 and y[i] < 0:
    coordinate.append(3)
  if x[i] > 0 and y[i] < 0:
    coordinate.append(4)

print(coordinate)

matrix = []  
for i in range(len(x)):
  matrix.append([x[i], y[i]])
          
print(matrix)

clf = svm.SVC(gamma = 0.0001, C = 100)
a = matrix[:-1]
b = coordinate[:-1]

clf.fit(a,b)
print("prediction: ", clf.predict([[-100, 50]]))
```
