```
# dataset
x1 = [1,1,2,3]
x2 = [0.5,1.5,1,1]
y = [2,3,4,4]
# parameter initialization
theta = [1,0.5,1]
```

```
# Hypothesis function
h = []
for i in range(len(x1)):
  a = theta[0] + theta[1] * x1[i] + theta[2] * x2[i]
  h.append(a)

print("Hypothesis function value is:", h)
```

```
# cost function

error_sum = 0
m = len(x1)
for i in range(m):
  error_sum = error_sum + ( h[i] - y[i] ) ** 2
  
j = ( 1/(2*m) ) * error_sum

print("Cost function value is: ", j)
```

```
# Gradient Descent

temp0 = 0
temp1 = 0
temp2 = 0
m = len(x1)

for i in range(m):
  temp0 = temp0 + ( h[i] - y[i] ) 
  temp1 = temp1 + ( h[i] - y[i] ) * x1[i]
  temp2 = temp2 + ( h[i] - y[i] ) * x2[i]

theta[0] = theta[0] - (0.01/m) * temp0
theta[1] = theta[1] - (0.01/m) * temp1
theta[2] = theta[2] - (0.01/m) * temp2

print("New value for theta(0) is: ", theta )
```
